```python
import setup_lab
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neural_network import MLPClassifier
from typing import List
from datasets.heart_dataset import HeartDataset
from common.plot import plot_pca_2d, plot_tsne_2d
```

https://www.kaggle.com/ronitf/heart-disease-uci

Attribute Information: 
> 1. age 
> 2. sex 
> 3. cp - chest pain type (4 values) 
> 4. resting blood pressure 
> 5. serum cholestoral in mg/dl 
> 6. fasting blood sugar > 120 mg/dl
> 7. restecg - resting electrocardiographic results (values 0,1,2)
> 8. thalach - maximum heart rate achieved 
> 9. exercise induced angina 
> 10. oldpeak = ST depression induced by exercise relative to rest 
> 11. slope - the slope of the peak exercise ST segment 
> 12. number of major vessels (0-3) colored by flourosopy 
> 13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

##### Preparing categorical features

```python
def prepair_categorical_features(df: pd.DataFrame, columns: List[str]):
    for col in columns:
        print(f"{col} values = ", sorted(df[col].unique()))

    dummy_cols = [pd.get_dummies(df[col], prefix=col) for col in columns]
    result = pd.concat([df, *dummy_cols], axis=1)
    result = result.drop(columns=columns)
    return result
```

```python
df = prepair_categorical_features(df, ['cp', 'thal', 'slope'])
df.head()
```

```python
x_df = df.drop(columns=[dataset.target_col])
y = df[dataset.target_col].values
```

##### Normalizing data

```python
x_df.values[:3]
```

```python
x = ((x_df - np.min(x_df)) / (np.max(x_df) - np.min(x_df))).values
```

```python
print(x[:3])
```

##### Training and Scores

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.45, random_state=25)
```

```python
print(x_train.shape)
print(y_train.shape)
print(sum(y_test == 0), sum(y_test == 1))
```

*Logistic Regression*

```python
lr = LogisticRegression(solver='lbfgs')
lr.fit(x_train, y_train)
print(f"Test Accuracy {lr.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, lr.predict(x_test), target_names=class_names))

cm_lr = confusion_matrix(y_test, lr.predict(x_test))
sns.heatmap(cm_lr, annot=True)
plt.show()

plot_pca_2d(x, y, model=lr, title=f"Logistic Regression")
plot_tsne_2d(x, y, title=f"Logistic Regression t-SNE")
```

*KNN*

```python
k = None
knn = None
for k_new in range(1, 30):
    knn_new = KNeighborsClassifier(n_neighbors=k_new)
    knn_new.fit(x_train, y_train)
    
    if knn is None or knn.score(x_test, y_test) < knn_new.score(x_test, y_test):
        k = k_new
        knn = knn_new

print(f"Best: {k}-NN Score: {knn.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, knn.predict(x_test), target_names=class_names))

cm_knn = confusion_matrix(y_test, knn.predict(x_test))
sns.heatmap(cm_knn, annot=True)
plt.show()

plot_pca_2d(x, y, model=knn, title=f"KNN")
```

*SVM*

```python
svm = SVC(gamma='auto', probability=True)
svm.fit(x_train, y_train)
print(f"Test Accuracy {svm.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, svm.predict(x_test), target_names=class_names))

cm_svm = confusion_matrix(y_test, svm.predict(x_test))
sns.heatmap(cm_svm, annot=True)
plt.show()

plot_pca_2d(x, y, model=svm, title=f"SVM")
```

*MLP*

```python
mlp = MLPClassifier(activation='relu', max_iter=1000)
mlp.fit(x_train, y_train)
print(f"Test Accuracy {mlp.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, mlp.predict(x_test), target_names=class_names))

cm_mlp = confusion_matrix(y_test, mlp.predict(x_test))
sns.heatmap(cm_mlp, annot=True)
plt.show()

plot_pca_2d(x, y, model=mlp, title=f"MLP Relu")
```

```python
mlp = MLPClassifier(activation='identity', max_iter=100)
mlp.fit(x_train, y_train)
print(f"Test Accuracy {mlp.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, mlp.predict(x_test), target_names=class_names))

cm_mlp = confusion_matrix(y_test, mlp.predict(x_test))
sns.heatmap(cm_mlp, annot=True)
plt.show()

plot_pca_2d(x, y, model=mlp, title=f"MLP Identity")
```

```python
mlp = MLPClassifier(activation='tanh', max_iter=2000)
mlp.fit(x_train, y_train)
print(f"Test Accuracy {mlp.score(x_test, y_test) * 100:.2f}%")
print(classification_report(y_test, mlp.predict(x_test), target_names=class_names))

cm_mlp = confusion_matrix(y_test, mlp.predict(x_test))
sns.heatmap(cm_mlp, annot=True)
plt.show()

plot_pca_2d(x, y, model=mlp, title=f"MLP Tanh")
```
